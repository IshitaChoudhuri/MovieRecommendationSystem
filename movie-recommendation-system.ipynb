<h5>Changing the dictionary form</h5>

```python
import ast

def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
result['genres'] = result['genres'].apply(convert)
```

```python
result['genres'].head()
```

```python
result['keywords'] = result['keywords'].apply(convert)
```

```python
result['keywords'].head()
```

```python
 def convert2(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
result['cast'] = result['cast'].apply(convert2)
```

```python
result.head()
```

```python
def director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
result['crew'] = result['crew'].apply(director)
```

```python
result.head()
```

```python
result['overview'] = result['overview'].apply(lambda x:x.split())
```

```python
result.head()
```

```python
result['genres'] = result['genres'].apply(lambda x:[i.replace(" ","") for i in x])
result['keywords'] = result['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
result['cast'] = result['cast'].apply(lambda x:[i.replace(" ","") for i in x])
result['crew'] = result['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
result.head()
```

```python
result['tags'] = result['overview'] + result['genres'] + result['keywords'] + result['cast'] + result['crew']
```

```python
result['tags'].head()
```

```python
new_df=result[['movie_id','title','tags']]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

<h5>Stemming</h5>
<p>the process of reducing a word to its word stem that affixes to suffixes and prefixes or the roots</p>

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
        
```

```python
new_df['tags'].apply(stem)
```

<h4>Vectorization</h4>
<p>converting input data from its raw format (text) into vectors of real numbers</p>

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors.shape
```

<h6>Determining the cosine distance</h6>

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim = cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = sim[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse = True, key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```
